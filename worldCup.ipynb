```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./worldCup.csv")
data = df
data.head()
```

```python
score = [data["home_score"].sum(), data["away_score"].sum()]

plt.pie(score, labels=["Home Score", "Away Score"], autopct="%1.1f%%", startangle=90)
plt.title("Percentage of goals for home and away teams")
plt.axis("equal")
plt.show()
```

```python
data["Host"].value_counts().plot(kind="bar", stacked=True, width=0.75)
plt.xticks(fontsize=8)
plt.title("Matches in each country")
plt.ylabel("Matches")
plt.xlabel("Host")
plt.show()
```

```python
print(
    f"Most matches was played in Germany by {data['Host'].value_counts().loc['Germany'].sum()} match"
)

print(
    f"Less matches was played in Uruguay by {data['Host'].value_counts().loc['Uruguay'].sum()} match"
)
```

```python
goals = data.groupby("Year")[["home_score", "away_score"]].sum()
goals.plot(kind="bar", stacked=True, width=0.75)
plt.title("Goals scored in world cup in each year")
plt.ylabel("Goals")
plt.xlabel("Year")
plt.show()
```

```python
print(f"Most goals scored in 2022: {goals.loc[2022].sum()} goal")

print(f"Less goals scored in 1930 and 1934: {goals.loc[1930].sum()} goal")
```

```python
grouped_data = data.groupby("Year")["Attendance"].sum()
grouped_data.plot(kind="bar", stacked=True, width=0.75)
plt.title("Attendance in world cup in each year")
plt.ylabel("Attendance")
plt.xlabel("Year")
plt.show()
```

```python
print(f"Most attendance in 1994: {grouped_data.loc[1994]} person")
print(f"Less attendance in 1934: {grouped_data.loc[1934]} person")
```

```python
team_counts = pd.concat([data["home_team"], data["away_team"]]).value_counts()
team_counts.plot(kind="bar", stacked=True, figsize=(14, 6),width=0.75)
plt.title("Matches for each team in world cup over all times")
plt.xticks(fontsize=7)
plt.ylabel("Matches")
plt.xlabel("Team")
plt.show()
```

```python
print(f"Most apperance in world cup for Brazil by {team_counts.loc['Brazil']} match")

print(
    f"Less apperance in world cup for Dutch East Indies by {team_counts.loc['Dutch East Indies']} match"
)
```

```python
stages = data.groupby("Round")[["home_score", "away_score"]].sum()
stages.plot(kind="bar", stacked=True, width=0.75)
plt.xticks(rotation=45, ha="right")
plt.title("Goals scored in each stage")
plt.ylabel("Goals")
plt.xlabel("Stage")
plt.show()
```
